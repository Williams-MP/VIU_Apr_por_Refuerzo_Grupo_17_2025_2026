````python
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4

env_name = 'Enduro-v0'
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n

print("Entorno:", env_name)
print("Número de acciones disponibles:", nb_actions)
print("Significado de las acciones:", env.unwrapped.get_action_meanings())
print("Formato de las observaciones:", env.observation_space)
````

````python
class AtariProcessor(Processor):

    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return float(np.clip(reward, -1., 1.))
````

1. Implementación de la red neuronal

````python
# 1. Implementación de la red neuronal
#
# Usamos la misma arquitectura convolucional propuesta por Mnih et al. (2015)
# para Atari: 3 capas convolucionales + 1 capa densa oculta + capa de salida
# lineal con un valor Q por acción. Es la misma red que en el ejemplo de
# Breakout del notebook de la sesión, adaptada al número de acciones de Enduro.

input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE


def build_model(nb_actions, input_shape):
    model = Sequential()
    if K.image_data_format() == 'channels_last':
        # (width, height, channels)
        model.add(Permute((2, 3, 1), input_shape=input_shape))
    elif K.image_data_format() == 'channels_first':
        # (channels, width, height)
        model.add(Permute((1, 2, 3), input_shape=input_shape))
    else:
        raise RuntimeError('Unknown image_dim_ordering.')

    model.add(Convolution2D(32, (8, 8), strides=(4, 4)))
    model.add(Activation('relu'))
    model.add(Convolution2D(64, (4, 4), strides=(2, 2)))
    model.add(Activation('relu'))
    model.add(Convolution2D(64, (3, 3), strides=(1, 1)))
    model.add(Activation('relu'))
    model.add(Flatten())
    model.add(Dense(512))
    model.add(Activation('relu'))
    model.add(Dense(nb_actions))
    model.add(Activation('linear'))
    return model


model = build_model(nb_actions, input_shape)
print(model.summary())
````

````python
# Función para cargar los pesos del último checkpoint disponible para realizar test

def load_weights_for_test(agent, proposal_id, weights_dir):
    """Carga el último checkpoint disponible para poder hacer test."""
    # Buscamos el último checkpoint disponible en la carpeta de pesos correspondiente a la práctica
    ckpt, step = find_latest_checkpoint(weights_dir, f'dqn_Enduro-v0_{proposal_id}')

    # Si se encuentra un archivo de pesos válido en el directorio
    if ckpt is not None:
        # Cargamos los pesos del checkpoint en el agente DQN
        agent.load_weights(ckpt)

        # Imprimimos un mensaje indicando que los pesos han sido cargados
        print(f"[{proposal_id}] Pesos cargados para test desde: {ckpt}  (step {step})")
    else:
        # Si no se encuentra ningún checkpoint, mostramos un mensaje de advertencia
        print(f"[{proposal_id}] Sin checkpoints. Entrena primero.")
````

````python
# Función para evaluar una propuesta cargando automáticamente su último checkpoint

def test_proposal(proposal_id, agents_registry, nb_episodes=NB_EPISODES):
    """
    Evaluamos una propuesta cargando automáticamente su último checkpoint.
    agents_registry: dict con la info de cada propuesta (dir, prefix, agente).
    """

    # Validamos si el identificador de la propuesta existe en nuestro diccionario de registro
    if proposal_id not in agents_registry:
        print(f"'{proposal_id}' no está en el registro. Opciones: {list(agents_registry.keys())}")
        return
    
    # Extraemos la información del agente correspondiente a la propuesta
    info = agents_registry[proposal_id]

    # Comprobamos si el agente está en memoria; si no lo está, indicamos que primero debe construirse
    if info['agent'] is None:
        print(f"[{proposal_id}] El agente no está en memoria todavía. Ejecuta primero su celda de construcción.")
        return

    # Buscamos el último checkpoint disponible en la carpeta de pesos correspondiente a la propuesta
    weights_filename, last_step = find_latest_checkpoint(info['dir'], info['prefix'])

    # Si no se encuentra ningún checkpoint, mostramos un mensaje de advertencia y salimos de la función
    if weights_filename is None:
        print(f"[{proposal_id}] Sin checkpoints en {info['dir']}. Entrena primero.")
        return

    # Cargamos los pesos entrenados en el agente antes de iniciar la evaluación
    print(f"[{proposal_id}] Cargando pesos desde: {weights_filename}  (step {last_step})")

    # Cargamos los pesos del último checkpoint en el agente y realizamos la evaluación
    info['agent'].load_weights(weights_filename)

    # Ejecutamos la evaluación del agente en el entorno durante un número especificado de episodios, sin visualización
    test_history = info['agent'].test(env, nb_episodes=nb_episodes, visualize=False)

    # Extraemos las recompensas obtenidas durante los episodios de prueba y calculamos la media y desviación estándar
    mean_reward = np.mean(test_history.history['episode_reward'])
    std_reward  = np.std(test_history.history['episode_reward'])

    # Mostramos los resultados finales por pantalla con formato de dos decimales.
    print(f"[{proposal_id}] Recompensa media en {nb_episodes} episodios: {mean_reward:.2f} (+/- {std_reward:.2f})")
    return test_history
````

##### Diagnóstico y diseño de las 3 propuestas de mejora

````python
# Heredamos de ModelIntervalCheckpoint para solucionar un problema de keras-rl2: su contador
# de guardado siempre empieza en 0. Al reanudar un entrenamiento de 10^6 pasos (por ejemplo, 
# tras una desconexión en el paso 800,000), el callback original sobreescribiría los primeros 
# pesos creados al inicio (100k, 200k...), destruyendo el historial.


class ResumableModelIntervalCheckpoint(ModelIntervalCheckpoint):

    def __init__(self, filepath, interval, step_offset=0, verbose=0): 
        # Llamamos al constructor de la clase base para inicializar los atributos heredados
        super().__init__(filepath, interval, verbose=verbose)

        # Guardamos el offset de pasos para poder reanudar el guardado desde un paso específico
        self.step_offset = step_offset

    def on_step_end(self, step, logs={}):
        # Incrementamos el contador de pasos totales y verificamos si es momento de guardar el modelo
        self.total_steps += 1

        # Si el número total de pasos no es múltiplo del intervalo, no hacemos nada y salimos
        if self.total_steps % self.interval != 0:
            return
        
        # Calculamos el paso absoluto sumando el offset al total de pasos y formateamos la ruta del archivo de pesos
        absolute_step = self.step_offset + self.total_steps
        
        # Formateamos la ruta del archivo de pesos usando el paso absoluto y los registros proporcionados
        filepath = self.filepath.format(step=absolute_step, **logs)
        
        # Si se ha configurado verbose, imprimimos un mensaje indicando que se está guardando el modelo
        if self.verbose > 0:
            print(f'Step {absolute_step}: guardando modelo en {filepath}')
        
        # Guardamos los pesos del modelo en la ruta especificada, sobrescribiendo cualquier archivo existente
        self.model.save_weights(filepath, overwrite=True)

# Función para encontrar el último checkpoint disponible en un directorio de pesos
def find_latest_checkpoint(weights_dir, prefix):
    """
    Busca en weights_dir el checkpoint con el step más alto que siga el
    patrón '{prefix}_weights_{step}.h5f.index', y devuelve una tupla
    (ruta_sin_extension, step) o (None, 0) si no hay ninguno.
    Útil para retomar automáticamente "desde el último punto guardado".
    """
    # Importamos los módulos necesarios para trabajar con expresiones regulares y búsqueda de archivos
    import re
    import glob

    # Construimos el patrón de búsqueda para los archivos de pesos en el directorio especificado
    pattern = os.path.join(weights_dir, f'{prefix}_weights_*.h5f.index')

    # Buscamos todos los archivos que coincidan con el patrón y almacenamos sus rutas en la lista candidates
    candidates = glob.glob(pattern)

    # Inicializamos la variable best_step con un valor negativo para asegurarnos de que cualquier step encontrado será mayor
    best_step = -1

    # Iteramos sobre los archivos encontrados para identificar el checkpoint con el step más alto
    best_path = None

    # Iteramos sobre los archivos encontrados para identificar el checkpoint con el step más alto
    for c in candidates:
        # Usamos una expresión regular para extraer el número de paso del nombre del archivo
        m = re.search(r'_weights_(\d+)\.h5f\.index$', c)

        # Si encontramos un número de paso válido, lo convertimos a entero y comparamos con el mejor paso encontrado hasta ahora
        if m:

            # Extraemos el número de paso del grupo capturado por la expresión regular y lo convertimos a entero
            step = int(m.group(1))

            # Si el paso actual es mayor que el mejor paso encontrado hasta ahora, actualizamos best_step y best_path
            if step > best_step:

                # Actualizamos best_step y best_path con el paso y la ruta del archivo actual
                best_step = step

                # Guardamos la ruta del archivo sin la extensión '.index' para poder cargar los pesos más tarde
                best_path = c[:-len('.index')]

    # Si no se encontró ningún checkpoint válido, devolvemos None y 0
    if best_path is None:
        return None, 0
    return best_path, best_step
````

````python

# Usar TRUE para entrenar, o FALSE para saltar el entrenamiento 
RUN_TRAINING_P5 = False          # <── GUARDIA PRINCIPAL

# Si se desea reanudar desde un checkpoint, establecer la ruta del checkpoint en RESUME_FROM_P5
if RUN_TRAINING_P5:

    # Si se encuentra un checkpoint válido, cargamos los pesos y ajustamos el número de pasos restantes para entrenar
    if RESUME_FROM_P5 is not None:
        print(f"Reanudando Propuesta 5 desde el checkpoint: {RESUME_FROM_P5}")

        # Cargamos los pesos del checkpoint en el agente DQN
        dqn_p5.load_weights(RESUME_FROM_P5)

        # Ajustamos el número de pasos restantes para entrenar, restando el offset de pasos ya completados
        NB_STEPS_REMAINING_P5 = max(0, NB_STEPS_TRAIN - STEP_OFFSET_P5)

    else:
        # Si no se encuentra ningún checkpoint, entrenamos desde cero y establecemos el número total de pasos de entrenamiento
        NB_STEPS_REMAINING_P5 = NB_STEPS_TRAIN

    # Guardamos los pesos finales de la Propuesta 5 en un archivo específico dentro de la carpeta de checkpoints correspondiente 
    weights_filename_p5 = os.path.join(WEIGHTS_DIR_P5, 'dqn_{}_p5_dueling_dqn_weights.h5f'.format(env_name))

    # Construimos la ruta del archivo de checkpoint para guardar los pesos periódicamente durante el entrenamiento
    checkpoint_filename_p5 = os.path.join(WEIGHTS_DIR_P5, 'dqn_' + env_name + '_p5_weights_{step}.h5f')

    # Construimos la ruta del archivo de log para registrar el progreso del entrenamiento en formato JSON
    log_filename_p5 = os.path.join(WEIGHTS_DIR_P5, 'dqn_{}_p5_log.json'.format(env_name))

    # Configuramos los callbacks para guardar checkpoints y registrar el progreso del entrenamiento
    callbacks_p5 = [ResumableModelIntervalCheckpoint(checkpoint_filename_p5, interval=100000,
                                                    step_offset=STEP_OFFSET_P5)]

    # Agregamos un callback adicional para registrar el progreso del entrenamiento en un archivo de log JSON cada 1000 pasos                                                    
    callbacks_p5 += [FileLogger(log_filename_p5, interval=1000)]

    # Entrenamos el agente DQN en el entorno durante el número de pasos restantes especificado, utilizando los callbacks configurados para guardar checkpoints y registrar el progreso del entrenamiento. La visualización se desactiva para acelerar el proceso de entrenamiento.
    history_p5 = dqn_p5.fit(env, callbacks=callbacks_p5, nb_steps=NB_STEPS_REMAINING_P5,
                            log_interval=10000, visualize=False)

    # Guardamos los pesos finales del modelo entrenado en el archivo especificado, sobrescribiendo cualquier archivo existente
    dqn_p5.save_weights(weights_filename_p5, overwrite=True)
    
    # Mostramos un mensaje indicando que los pesos finales de la Propuesta 5 han sido guardados en Google Drive, junto con la ruta del archivo de pesos.
    print("Pesos finales de la Propuesta 5 guardados en Drive:", weights_filename_p5)
    
    # Mostramos el contenido de la carpeta de checkpoints de la Propuesta 5 para verificar que los archivos se han guardado correctamente.
    print("Contenido de la carpeta:", os.listdir(WEIGHTS_DIR_P5))

else:
    print("[P5] Entrenamiento omitido (RUN_TRAINING_P5=False).")
````
